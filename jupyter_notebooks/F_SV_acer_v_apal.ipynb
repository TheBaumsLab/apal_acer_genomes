```python
variants = pd.read_csv("../AP_AC_mum.SVs_all.vcf", 
                       comment = "#", sep = "\t", header = None)
acer_support = pd.read_csv("../recAllRatesLoess_Acroporacervicornis_FEMALE_2023-12-14.txt", sep = " ")
```

```python
variants["SVLEN"] = variants[7].str.split(";", n = 2, expand = True)[1].str.replace("SVLEN=", "")
```

```python
inv = variants[variants[7].str.contains("INV")][variants[6] == "PASS"].reset_index(drop=True)
inv["SVLEN"] = inv["SVLEN"].astype(int)
```

```python
inv["mkr_support"] = np.nan
for i in range(len(inv)):
    inv["mkr_support"][i] = len(acer_support[acer_support["map"] == inv[0][i]]
        [acer_support["phys"] > inv[1][i]][acer_support["phys"] < ( inv[1][i] + inv["SVLEN"][i] )])
```

```python
tra = variants[variants[7].str.contains("TRA")][variants[6] == "PASS"].reset_index(drop=True)
tra["SVLEN"] = tra["SVLEN"].astype(int)
```

```python
tra["mkr_support"] = np.nan
for i in range(len(tra)):
    tra["mkr_support"][i] = len(acer_support[acer_support["map"] == tra[0][i]]
        [acer_support["phys"] > tra[1][i]][acer_support["phys"] < ( tra[1][i] + tra["SVLEN"][i] )])
```
